```python
import pandas as pd
import os
from statsmodels.stats.inter_rater import fleiss_kappa, aggregate_raters



df_yuma = pd.read_csv(os.path.join(path, 'labeled_data_yuma.csv')).iloc[0:100]
df_rehan = pd.read_csv(os.path.join(path, 'labaled_data_rehan.csv')).iloc[0:100]
df_ameng = pd.read_csv(os.path.join(path, 'labeled_data_ameng.csv')).iloc[0:100]


df_kappa = pd.DataFrame({
    'Yuma': df_yuma['Stance'],
    'Rehan': df_rehan['Stance'],
    'Ameng': df_ameng['Stance']
})

df_kappa = df_kappa.astype(str).apply(lambda x: x.str.strip().str.lower())


data_agregat,categories = aggregate_raters(df_kappa.values)
kappa = fleiss_kappa(data_agregat, method='fleiss')

print(f"Skor Fleiss' Kappa: {kappa:.4f}")
```

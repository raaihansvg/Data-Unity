```python
import pandas as pd

data = pd.read_csv("/content/drive/MyDrive/Gemastik/DataBelumFinal(sebelum kappa)/PunyaRaihan.csv",header=4)
```

```python
data.columns = data.iloc[0]
data = data[1:]
data = data.reset_index(drop=True)
```

```python
data.columns
```

```python
data.head(10)
```

```python
data.shape
```

#Bagian 1


```python
batchSize = 100

start = 0
end = start + batchSize

dataBatch = data.iloc[start:end].reset_index(drop=True)

labels = []

mapping = {
    "1": "supportive",
    "2": "VictimBlaming",
    "3": "Neutral",
    "4": "Advocacy"
}

for i, row in dataBatch.iterrows():
    print("\n=================")
    print(f"Data ke-{start + i}")
    print("Text:", row['Teks Komentar'])

    print("\n1 = supportive")
    print("2 = VictimBlaming")
    print("3 = Neutral")
    print("4 = Advocacy")

    label = input("Masukan Angka: ")

    labels.append(mapping.get(label, "Neutral"))
```

```python
dataBatch.columns
```

```python
dataBatch['Stance'] = labels
dataBatch.to_csv(f'/content/drive/MyDrive/Gemastik/DataBelumFinal(sebelum kappa)/labaled_{start}_{end}.csv',index=False)
```

```python
import os
path = "/content/drive/MyDrive/Gemastik"
os.listdir(path)
```

```python
import pandas as pd
import glob

files = glob.glob(path + "/*.csv")

df_list= []

for x in files:
  df = pd.read_csv(x)
  df = df[['text']]
  df_list.append(df)


DataFinalUNY = pd.concat(df_list,ignore_index=True)
print("total data: ", len(DataFinalUNY))
```

```python
DataFinalUNY.drop_duplicates(subset='text',inplace=True)
print(len(DataFinalUNY))
```

```python
DataFinalUNY['stance'] = ''
DataFinalUNY.to_csv(path + '/DataFinalUNY.csv',index=False)
DataFinalUNY.head()
```
